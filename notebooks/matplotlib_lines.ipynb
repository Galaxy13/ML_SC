```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,color='#650dbd',label='X',lw=2,ls='--',marker='o',ms=10) # RGB Hex Mode

axes.plot(x, x + 2, color='red',label='X + 2',linewidth=2,linestyle='--',marker='s',markersize=10,markerfacecolor='red',markeredgewidth=3,markeredgecolor='blue')
# lines[0].set_dashes([5,2,1,2]) # Size of dash, size of white space, size of next dash, size of white space...
axes.legend()
```
