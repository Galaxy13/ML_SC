```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df['total'] = df["TV"] + df["radio"] + df["newspaper"]
```

```python
sns.regplot(data=df, x='total', y='sales')
```

```python
X = df['total']
Y = df['sales']
```

```python
beta_coef = np.polyfit(X, Y, deg=1)
```

```python
potential_spend = np.linspace(0, 500, 100)
predicted_sales = beta_coef[0] * potential_spend + beta_coef[1]
```

```python
df2 = pd.DataFrame(columns=[potential_spend, predicted_sales])
```

```python
sns.scatterplot(data=df, x='total', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```

```python
spend = 200
predicted_sale = beta_coef[0] * spend + beta_coef[1]
predicted_sale
```

```python
beta_coef_deg3 = np.polyfit(X, Y, deg=3)
```

```python
pot_spend = np.linspace(0, 500, 100)

pred_sales = pot_spend ** 3 * beta_coef_deg3[0] + pot_spend ** 2 * beta_coef_deg3[1] + pot_spend * beta_coef_deg3[2] +\
                beta_coef_deg3[3]
```

```python
sns.scatterplot(data=df, x='total', y='sales')
plt.plot(pot_spend, pred_sales, color='red')
```
