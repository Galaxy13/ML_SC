```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Home\PycharmProjects\UDEMY_ML_NEW\data\StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df, y='reading score', x='parental level of education',
            hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',
               hue='test preparation course',split=True,inner='stick',bw=0.4)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df, x='math score',size=2,y='gender',
              hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df,
              hue='gender')
```
