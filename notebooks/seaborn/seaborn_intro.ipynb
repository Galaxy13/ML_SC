```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Home\PycharmProjects\UDEMY_ML_NEW\data\dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',
                palette='Dark2',size='salary',s=200,alpha=0.5,style='level of education')
plt.savefig('my_plot.png')
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df, x='salary',bins=50,color='red',edgecolor='blue',linewidth=1,kde=True,rug=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x='salary')
```
