```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 12), dpi=200)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][0].set_xlabel('X')
axes[0][0].set_xlim(5,10)
fig.suptitle('Figure')

fig.subplots_adjust(wspace=0.8, hspace=0.3)
fig.savefig('Mat_adv', bbox_inches='tight')
```
