```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
u_anod = np.linspace(10, 90, num=9)
u_anod = np.insert(u_anod, 0, 1)
u_anod
```

```python
lab_frame_1 = pd.DataFrame(data=u_anod, columns=['Ua'])
lab_frame_1['S/Ia'] = 3/2 * lab_frame_1['Ua'] ** (-1/2)
lab_frame_1.to_excel('Крутизна.xlsx')
```

```python
plt.figure(dpi=300)
sns.set_style(style='whitegrid')
plot = sns.lineplot(data=lab_frame_1, x='Ua', y='S/Ia', color='black', linewidth=1, marker='o', dashes=True,
                    palette='black')
plot.set(xlabel='Напряжение (Ua), В', ylabel='Удельная крутизна по току, [мА/В]')
plot.set_title('Зависимость удельной крутизны по току от анодного напряжения')
fig = plot.get_figure()
fig.savefig('plot_lab_1-2.2(1).png')
```
