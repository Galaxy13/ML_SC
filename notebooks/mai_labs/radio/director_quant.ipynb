```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
elem = np.array([3, 5, 7, 10, 11, 16, 20, 24, 30])
coef = np.array([5.5, 8.7, 10, 11.5, 12, 13.5, 16.5, 17, 18.5])
```

```python
lab_frame = pd.DataFrame(data=elem, columns=['elem'])
lab_frame['coef'] = coef
```

```python
plt.figure(dpi=300)
sns.set_style('whitegrid')
plot = sns.lineplot(data=lab_frame, x='elem', y='coef')
plot.set(xlabel=' ', ylabel=' , ')
fig = plot.get_figure()
fig.savefig('1.png')
```
