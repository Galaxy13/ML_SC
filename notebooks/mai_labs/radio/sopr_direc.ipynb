```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sopr = np.array([10, 21, 32, 43, 54, 65, 76, 87, 98, 109, 120])
coef = np.array([1.2, 1.3, 1.4, 1.54, 1.65, 1.81, 1.88, 2.0, 2.2, 2.58, 2.7])
```

```python
lab_frame1 = pd.DataFrame(data=sopr, columns=['sopr'])
lab_frame1['coef'] = coef
```

```python
plt.figure(dpi=300)
sns.set_style('whitegrid')
plot = sns.lineplot(data=lab_frame1, x='sopr', y='coef')
plot.set(xlabel='X00, ', ylabel='Em(0)/Em(180)')
plt.ylim(0,3)
fig = plot.get_figure()
fig.savefig('1.png')
```

```python
lab_frame1.to_excel('sopr1.xlsx')
```
