```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
u_anod = np.linspace(10, 100, num=10)
u_anod = np.insert(u_anod, 0, 1)
u_anod
```

```python
lab_frame = pd.DataFrame(data=u_anod, columns=['Ua'])
```

```python
lab_frame['Ua'][0] = 140
lab_frame['t'] = 5.06 * 10 ** (-6) * 0.0025 / lab_frame['Ua'] ** (1/2)
lab_frame['nanosec_t'] = lab_frame['t'] * 10 ** 9
lab_frame
```

```python
plt.figure(dpi=300)
sns.set_style(style='darkgrid')
plot = sns.lineplot(data=lab_frame, x='Ua', y='nanosec_t', color='black', linewidth=1, marker='o', dashes=True,
                    palette='black')
plot.set(xlabel='Напряжение (Ua), В', ylabel='Время пролета, нс')
plot.set_title('Зависимость времени пролёта от анодного напряжения')
fig = plot.get_figure()
fig.savefig('plot_lab_1-2.png')
```
