```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
u_anod = np.linspace(0, 100, num=11)
```

```python
lab_frame = pd.DataFrame(data=u_anod, columns=['Ua'])
```

```python
lab_frame['Ua'][4] = 45
lab_frame['Ua'][3] = 35
lab_frame['Ia'] = (lab_frame['Ua'] ** (3/2)) * (2.33 / 100000 * 0.00015 / (0.0019 ** (3/2)))
lab_frame['Ia_mil'] = lab_frame['Ia'] * 1000
lab_frame
# lab_frame['Ua'][5] = 45.0
# lab_frame['Ia'][5] = (lab_frame['Ua'][5] ** (3/2)) * (2.33 / 100000 * 0.00015 / (0.0019 ** (3/2)))
# lab_frame['Ia_mil'][5] = lab_frame['Ia'][5] * 1000
# lab_frame['Ua'][10] = 44.9
# lab_frame['Ia'][10] = (lab_frame['Ua'][10] ** (3/2)) * (2.33 / 100000 * 0.00015 / (0.0019 ** (3/2)))
# lab_frame['Ia_mil'][10] = lab_frame['Ia'][10] * 1000
# lab_frame
```

```python
plt.figure(dpi=300)
sns.set_style(style='darkgrid')
plot = sns.lineplot(data=lab_frame, x='Ua', y='Ia_mil', color='blue', linewidth=1, marker='o', dashes=True)
plot.set(xlabel=' (Ua), ', ylabel=' (Ia), ')
plot.set_title('  ')
fig = plot.get_figure()
plt.plot([45, 45], [0,12.73], linewidth=1, color='black', linestyle='--')
plt.plot([45,0], [12.73,12.73], lw=1, color='black',linestyle='--')
plt.plot([15,45],[0,12.73], linewidth=1, color='red', linestyle='-')
plt.plot([0,45],[0,12.73], linewidth=1, color='green', linestyle='-')
# plt.plot([74,100], [27, 27], lw=1, color='black', linestyle='--')
# plt.plot([100,100], [0, 27], lw=1, color='black', linestyle='-')
# plt.plot([45,0], [0, 46.7], linewidth=1, color='red')
# plt.plot([80,60], [0, 20], linewidth=1, color='red')
# plt.plot([100,74], [0, 27], linewidth=1, color='red')
# plt.plot([0,50], [0,14.92], lw=1, color='black')

# sns.lineplot(data=lab_frame, x='Ua', y='Ia_mil_new', color='green', lw=1, dashes=True)
```
